# 문제 PY45-1010

하나의 문자열을 입력받고, 소문자를 대문자로 변환하는 프로그램을 작성하시오.

- 조건
    
    특수기호를 제거하고, 소문자를 대문자로 변환

```python
str = input()
result = ""
for i in str:
  if i.isalnum():
    if i.islower():
      result += i.upper()
    else:
      result += i
print(result)

result = "".join([i.upper() for i in str if i.isalnum()])
print(result)
```

# 문제 PY45-1011

### 문제

입력받은 문자열에서 알파벳 대문자와 소문자의 개수를 각각 세어서 출력하는 프로그램을 작성하시오.

---

### 조건

1. 입력으로 주어지는 문자열은 공백을 포함할 수 있다.
2. 문자열의 길이는 1 이상 1,000 이하이다.
3. 대문자와 소문자 이외의 문자는 무시한다.
4. 대문자와 소문자의 개수를 각각 출력한다.

```python
str = input()
lowCount = 0
upCount = 0
for i in str:
  if i.islower():
    lowCount += 1
  if i.isupper():
    upCount += 1
print(f"Capital: {upCount}, Small: {lowCount}")
```

# 문제 PY45-1014

문자열이 주어졌을 때, 모든 소문자를 대문자로, 대문자를 소문자로 바꾸는 프로그램을 작성하세요.

**조건**

- 입력 문자열은 알파벳으로만 이루어져 있으며, 공백은 없습니다.
- 문자열의 길이는 1 이상 100 이하입니다.
- `str.swapcase()` 함수를 사용 x

```python
str = input()
result = ""
for i in str:
  if i.islower():
    result += i.upper()
  elif i.isupper():
    result += i.lower()
  else:
    result += i
result
```

# 문제 PY45-1015

문자열과 제거할 문자가 주어졌을 때, 해당 문자를 제거한 문자열을 출력하세요.

**조건**

- 입력 문자열의 길이는 1 이상 100 이하입니다.
- 제거할 문자는 한 글자이며, 반드시 문자열 안에 포함되어 있습니다.
- (방법1) `str.replace()`
- (방법2) 리스트 컴프리헨션을 사용할 수 있습니다.

```python
# (방법1) str.replace()
str = input().split(" ")
result1 = (str[0].replace(str[1], "")).strip()
# (방법2) 리스트 컴프리헨션을 사용할 수 있습니다.
result2 = "".join([i.strip() for i in str[0] if i != str[1]])

print(result1)
print(result2)
```

# 문제 PY45-1017

문자열이 주어졌을 때, 각 단어의 순서는 유지하되 단어 내부의 알파벳을 뒤집은 문자열을 출력하세요.

**조건**

- 입력 문자열은 공백으로 구분된 단어들로 구성됩니다.
- 방법1) 슬라이스연산자
- 방법2) reversed() 사용

```python
str = input()
# 방법1) 슬라이스연산자
result1 = str[::-1]
# 방법2) reversed() 사용
result2 = ''.join(reversed(str))
print(result1)
print(result2)
```
