# 문제 NP12-0001

다음과 3차원 배열 np1 을 만들고, axis 에 따른 합을 구하는 프로그램의 결과를 그리시오.


```python
np2 = np1.sum(axis=0)
np3 = np1.sum(axis=1)
np4 = np1.sum(axis=2)
```



````python
import numpy as np
np1 = np.arange(0, 24).reshape(2, 3, 4)
print(np1)

np2 = np1.sum(axis=0)
print("\naxis=0 블록끼리 행 기준")
print(np2)

np3 = np1.sum(axis=1)
print("\naxis=1 블록 내 행끼리")
print(np3)

np4 = np1.sum(axis=2)
print("\naxis=2 블록 내 행의 열끼리")
print(np4)
````

# 문제 NP12-0002

임의값으로 3x6 배열을 만들고, 최대 최소값을 출력하는 프로그램을 작성하시오.

- 조건
    - np.random.random() 함수 사용
    - **소숫점2째 자리까지 표현**
    - **행별** 최소, 최대값을 출력한다. (아래 표 참조)
    - 반복문 사용x
    - 리스트 사용 x

````python
import numpy as np
np1 = np.round(np.random.random((1, 3, 6)), 2)
print(np1)

# axis=2 : 블록 내 행의 열끼리
print('max:', np.max(np1, axis=2))
print('min:', np.min(np1, axis=2))
````

# 문제 NP12-0003

임의값으로 3x6 배열을 만들고, 최대 최소값을 출력하는 프로그램을 작성하시오.

- 조건
    - np.random.random() 함수 사용
    - **소숫점2째 자리까지 표현**
    - **열별** 최소, 최대값을 출력한다. (아래 표 참조)
    - 반복문 사용x
    - 리스트 사용 x

````python
import numpy as np

np1 = np.random.random((1,3,6))
np1_rounded = np.round(np1, 2)
print(np1_rounded)

# axis=1: 블록 내의 행끼리
print('max:', np.max(np1_rounded, axis=1))
print('min:', np.min(np1_rounded, axis=1))
````
