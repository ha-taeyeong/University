# 문제 PLT10-0010

두 직선의 교점을 구하시오.

- 첫 번째 직선의 방정식은 $y=−3x+7$
- 두 번째 직선의 방정식은 $y=2x−3$
- 두 직선을 그래프로 나타내고, 교점에 점을 찍어 표시하시오.
- 연립방정식으로 해를 구하고, `np.linalg.solve()`를 이용하시오.

**조건**

- 첫 번째 직선: 기울기 -3, y절편 7
- 두 번째 직선: 기울기 2, y절편 -3
- x의 범위는 -5부터 5까지 0.1 간격
- 교점 좌표를 출력하고, 해당 좌표를 그래프에 점으로 표시

````python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[1, 3], [1, -2]])
B = np.array([7, -3])
intersection = np.linalg.solve(A, B)
print(intersection)

x = np.arange(-5, 5.1, 0.1)
y1 = -3 * x + 7
y2 = 2 * x - 3
x_cross, y_cross = 2.0, 1.0

plt.plot(x, y1, label='y = -3x + 7')
plt.plot(x, y2, label='y = 2x - 3')
plt.scatter(x_cross, y_cross, color='red', zorder=5, label='intersection')
plt.legend()
plt.grid(True)
plt.show()
````

# 문제 PLT10-0011

다음 연립방정식을 이용하여 두 직선의 교점을 구하시오.

- 첫 번째 직선의 방정식은 y=0.5x+4
- 두 번째 직선의 방정식은 y=−x+1
- 두 직선을 같은 그래프에 그리고 교점에 점을 찍으시오.

**조건**

- 첫 번째 직선: 기울기 0.5, y절편 4
- 두 번째 직선: 기울기 -1, y절편 1
- x의 범위는 -5부터 5까지 0.1 간격
- `np.linalg.solve()`를 사용하여 해를 구하고 출력

````python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[1, -0.5], [1, 1]])
B = np.array([4, 1])
intersection = np.linalg.solve(A, B)
print(intersection)

x = np.arange(-5, 5.1, 0.1)
y1 = 0.5 * x + 4
y2 = -1 * x + 1
x_cross, y_cross = -2.0, 3.0

plt.plot(x, y1, label='y=0.5x+4')
plt.plot(x, y2, label='y=-x+1')
plt.scatter(x_cross, y_cross, color='red', zorder=5, label='intersection')
plt.legend()
plt.grid(True)
plt.show()
````

# 문제 PLT11-0002

다음 그림과 같이 2개의 직선의 방정식을 그리고 교점을 찾으시오.

- 조건
    - 직선1은 (0,3) (-4,-5) , 직선2는 (0,11), (-4,-5) 를 지난다.
        - 1차 방정식을 만들어서 각각 그린다.
    - 교점은 행렬을 이용(선형대수 np.linalg) 한다.

````python
import numpy as np
import matplotlib.pyplot as plt

# y=2x+3, y=4x+11
A = np.array([[-2, 1], [-4, 1]])
B = np.array([3, 11])
intersection = np.linalg.solve(A, B)
x_cross, y_cross = intersection

x = np.linspace(-5, 5, 100)
y1 = 2 * x + 3
y2 = 4 * x + 11

plt.plot(x, y1, label="y=2x+3")
plt.plot(x, y2, label="y=4x+11")
plt.scatter(x_cross, y_cross, color="red", zorder=5, label="intersection")
plt.legend()
plt.grid(True)
plt.show()
````

# 문제 PLT11-0003

다음 그림과 같이 2개의 삼각함수를 그리시오.

- 조건
    - np.sin() 함수를 사용한다.
    - np.cos() 함수를 사용한다.
- 입력
    - 없음

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi, 100)
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x, sin_y, label="sin(x)")
plt.plot(x, cos_y, label="cos(x)")
plt.legend()
plt.show()
````

# 문제 PLT11-0004

다음 그림과 같이 sin() 함수를 그리시오.

- 조건
    - numpy 를 사용하지 않는다.
    - math.sin() 함수를 사용한다.
- 입력
    - 없음

````python
import math
import matplotlib.pyplot as plt

num_points =100
start = 0
end = 4*math.pi
step = (end - start) / (num_points-1)
x_values = [start + i * step for i in range(num_points)]
sin_values = [math.sin(x) for x in x_values]
cos_values = [math.cos(x) for x in x_values]
plt.plot(x_values, sin_values, linestyle='dotted')
plt.plot(x_values, cos_values, linestyle='dotted')
plt.show()
````

# 문제 PLT11-0005
다음 그림과 같이 sin(), cos() 함수를 그리시오.
- 조건
 - plot() 함수의 data 파라메터를 이용한다.
   - 딕셔너리를 이용해서 독립변수와 종속변수 테이블을 구성한다.


```python
plt.plot('x','sin(x)', data=dic1)
plt.plot('x','cos(x)', data=dic1)
```



````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi, 100)

dic1 = {
    'x': x,
    'sin(x)': np.sin(x),
    'cos(x)': np.cos(x)
}
plt.plot('x', 'sin(x)', data=dic1, label='sin(x)')
plt.plot('x', 'cos(x)', data=dic1, label='cos(x)')
plt.legend()
plt.show()
````
