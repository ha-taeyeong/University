### 문제 1:

1부터 20까지의 숫자 중에서 소수만 리스트로 만드세요.

- 조건 :
    - 두가지 풀이방법 모두 작성하시오.
    - 방법1) LC 사용
    - 방법2) LC x  (사용자정의함수 사용)

````python
# 방법1) LC 사용
result1 = [x for x in range(2, 21) if all(x % y != 0 for y in range(2, x))]
print(result1)

# 방법2) LC x (사용자정의함수 사용)
def is_prime(x):
  if x < 2:
    return False
  for y in range(2, x):
    if x % y == 0:
      return False
  else:
    return True

result2 = []
for x in range(2, 21):
  if is_prime(x):
    result2.append(x)
print(result2)
````

###문제 3:
주어진 리스트에서 중복된 숫자를 제거하고, 중복이 없는 리스트를 만드세요.
- 조건 :
 - 두가지 풀이방법 모두 작성하시오.
 - 방법1) LC 사용
 - 방법2) LC x  (사용자정의함수 사용)
 - 방법3) LC 사용 + 슬라이싱 이용

```
numbers = [1, 2, 2, 3, 4, 4, 5]
```

````python
numbers = [1, 2, 2, 3, 4, 4, 5]
# 방법1) LC 사용
result1 = []
[result1.append(x) for x in numbers if x not in result1]
print(result1)

# 방법2) LC x (사용자정의함수 사용)
def remove_duplicates(numbers):
  result = []
  for x in numbers:
    if x not in result:
      result.append(x)
  return result
result2 = remove_duplicates(numbers)
print(result2)

# 방법3) LC 사용 + 슬라이싱 이용
result3 = [x for i, x in enumerate(numbers) if x not in numbers[:i]]
print(result3)
````
