# 문제 PY93-0001

0부터 9까지 자연수 10개를 **랜덤 원소**로 가지는 리스트를 3가지 방법으로 각각 만들어보자

- 조건
    - 원소의 순서는 Random
    - **원소는 중복되지 않도록 한다.**
- 방법
    - 방법 1: 반복문 1회 사용
    - 방법 2: 반복문 2회 이상 사용
    - 방법 3: random.sample() 함수 사용
- 출력결과 : [3, 1, 8, 2, 9, 5, 0, 7, 4, 6]

````python
# 방법 1: 반복문 1회 사용
import random

nums = list(range(10))
random.shuffle(nums)
print(nums)
````

````python
# 방법 2: 반복문 2회 이상 사용
import random

nums = []
pool = list(range(10))
for _ in range(10): # 10번 반복
  idx = random.randint(0, len(pool) - 1)
  nums.append(pool.pop(idx)) # pool에서 랜덤하게 꺼내서 추가
print(nums)
````

````python
# 방법 3: random.sample() 함수 사용
import random

nums = random.sample(range(10), 10)
print(nums)
````

#문제 PY93-0002
주사위 1개와 동전1개를 10번 중복가능 추출한 결과를 출력하시오
- 조건
 - 방법1) randint() 함수를 사용
 - 방법2) choices() 함수를 사용
- 중복가능
- 출력예
  ```python
    [('H', 6),
    ('H', 5),
    ('H', 4),
    ('T', 6),
    ('T', 4),
    ('H', 4),
    ('H', 4),
    ('T', 1),
    ('T', 6),
    ('T', 4)]
  ```

````python
# 방법 1: randint() 함수 사용
import random
result = []
for _ in range(10):
  coin = 'H' if random.randint(0, 1) == 0 else 'T'
  dice = random.randint(1, 6)
  result.append((coin, dice))
print(result)
````

````python
# 방법 2: choices() 함수 사용
import random

coins = random.choices(['H', 'T'], k=10)
dices = random.choices(list(range(1,7)), k=10)
result = list(zip(coins, dices))
print(result)
````
