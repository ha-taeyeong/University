# 문제 PD13-0004

1부터 10까지 자연수 리스트를 만들고, 짝수를 출력하는 프로그램을 작성하시오.

- 조건
    - 풀이1) map 함수 이용
    - 풀이2) filter 함수사용
    - 풀이3) numpy TF 마스크이용
    - 풀이4) numpy 인덱스이용
    - 풀이5) pandas TF 마스크사용
    
- 출력예
    - [2, 4, 6, 8, 10]



```python
# 풀이1) map 함수 이용
numbers = list(range(1, 11))
is_even = lambda x: x if x % 2 == 0 else None
mapped = list(map(is_even, numbers))
result_map = [x for x in mapped if x is not None]
print(result_map)
```

```python
# 풀이2) filter 함수사용
numbers = list(range(1, 11))
result_filter = list(filter(lambda x: x%2 == 0, numbers))
print(result_filter)
```

```python
# 풀이3) numpy TF 마스크이용
import numpy as np
numbers_np = np.arange(1, 11)
mask = (numbers_np % 2 == 0)
result_np_mask = numbers_np[mask].tolist()
print(result_np_mask)
```

```python
# 풀이4) numpy 인덱스이용
import numpy as np
numbers_np = np.arange(1, 11)
idx = np.where(numbers_np % 2 == 0)
result_np_idx = numbers_np[idx].tolist()
print(result_np_idx)
```

```python
# 풀이5) pandas TF 마스크사용
import pandas as pd
numbers_pd = pd.Series(range(1, 11))
result_pd_mask = numbers_pd[numbers_pd % 2 == 0].tolist()
print(result_pd_mask)
```
